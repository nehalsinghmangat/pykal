# Overview

This tutorial will guide the reader through the basic usage of the `pykal_core` library. 

This tutorial will begin with a review of the basic feedback system; then we will discuss the implementation of the feedback system in `pykal_core` using object-oriented progamming; finally, the implementation of each feedback system class will be discussed and examples of their usage will be shown. When appropriate, the theoretical considerations behind certain design decisions will be discussed. 

Upon finishing this tutorial, the reader will be able to model basic control systems using the pykal_core library, as shown in tutorial 1. [link] The reader will also be able to model a basic robot control system,  as shown in tutorial 1 [link].
